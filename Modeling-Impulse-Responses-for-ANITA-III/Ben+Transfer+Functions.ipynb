```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.fft as fft
import scipy as sp
import scipy.signal

%matplotlib inline
```

```python
data = np.loadtxt('tf08TH.txt')
time = data[:,0]
volt = data[:,1]
plt.plot(time, volt)

plt.title('Impulse Response of 08TH Channel', fontsize=14)
plt.xlabel('Time (ns)', fontsize = 14)
plt.ylabel('Amplitude (V)', fontsize =14)
```

```python
N=len(time)
dt=time[1]-time[0]

f = fft.fft(volt)
fshift = fft.fftshift(f)
freq = fft.fftfreq(N, dt)



v_squared=abs(fshift*fshift)
Power=v_squared/50
mW=Power*10**3



delta_f=(freq[2]-freq[1])*1e9
mWHz1=mW/delta_f 



for i in range(len(freq)):
    print(fshift[i],',')



dBm1=10*np.log10(mWHz1)

plt.plot(fft.fftshift(freq), dBm1)
plt.xlim([0,1.35])

plt.title('Spectrum of 08TH Channel', fontsize=14)
plt.xlabel('Frequnecy (GHz)', fontsize = 14)
plt.ylabel('dB(mW/Hz)', fontsize =14)
```
